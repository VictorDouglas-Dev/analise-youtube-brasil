```python
import pandas as pd

# Carregar o dataset que você limpou no seu script
df = pd.read_csv('../data/processed/top_youtube_brasil_limpo.csv')

# Exibir as 5 primeiras linhas para conferência
df.head()
```

```python
import matplotlib.pyplot as plt

# Conta quantos canais existem em cada porte
contagem_porte = df['Porte_do_Canal'].value_counts()

# Cria o gráfico de barras
contagem_porte.plot(kind='bar', color='skyblue', figsize=(8, 5))

# Adiciona títulos e labels para ficar profissional
plt.title('Distribuição de Canais por Porte')
plt.xlabel('Porte do Canal')
plt.ylabel('Quantidade de Canais')
plt.xticks(rotation=0)
plt.show()
```

```python
# Ver os 5 canais com maior "eficiência" (Views por Inscrito)
print(df.nlargest(5, 'Views_por_Inscrito')[['Channel Name', 'Views_por_Inscrito']])
```

```python
# Ver os 5 canais com menor "eficiência"
print(df.nsmallest(5, 'Views_por_Inscrito')[['Channel Name', 'Views_por_Inscrito']])
```
